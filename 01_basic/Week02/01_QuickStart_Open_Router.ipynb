```python
# Check the keys

import os
open_router_api_key = os.getenv('OPEN_ROUTER_API_KEY')

if open_router_api_key:
    print(f"Open router API Key exists and begins {open_router_api_key[:8]}")
else:
    print("Open router API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
import os
from openai import OpenAI

client = OpenAI(base_url="https://openrouter.ai/api/v1",api_key=open_router_api_key)
model_name = "deepseek/deepseek-r1-0528:free"    
```

```python
user_prompt = "ขอวิธีทำข้าวผัดกระเพราไก่หน่อยครับ"

completion = client.chat.completions.create(
    model=model_name ,
    messages=[
        {"role": "user", "content": user_prompt},
    ],
    
    temperature=0.8,
   
)

response = completion.choices[0].message.content
print(response)
```

# Multi model with API

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# Load environment variables
load_dotenv()
```

```python
import os
from openai import OpenAI

client = OpenAI(base_url="https://openrouter.ai/api/v1",api_key= os.getenv('OPEN_ROUTER_API_KEY'))

```

```python
import openai


models = [
    "google/gemini-2.5-flash",
    "anthropic/claude-sonnet-4",
     "deepseek/deepseek-r1-distill-qwen-7b",
     "meta-llama/llama-4-maverick"
    
]


models
```

```python

user_prompt = "ทำไมการนอนหลับให้เพียงพอจึงสำคัญต่อสุขภาพจิต?"

for model_name in models:
    print(f"\n\n### Model: {model_name} ###\n")
    try:

        completion = client.chat.completions.create(
        model=model_name ,
        messages=[
            {"role": "user", "content": user_prompt},
        ],temperature=0.0)

        response = completion.choices[0].message.content

        print(f"Response from {model_name}:\n{response}\n")


    except Exception as e:
      print(f"Error with model {model_name}: {e}")
```
