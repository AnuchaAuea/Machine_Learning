# Basic

**พารามิเตอร์สำคัญใน `openai_client.chat.completions.create`**

| พารามิเตอร์   | ใช้ทำอะไร                                             | ค่า-ยิ่งมาก/น้อยส่งผลอย่างไร                                                                                                                                                  | ค่าปกติ (ถ้าไม่กำหนด)                    |
| ------------- | ----------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------- |
| `temperature` | ควบคุม “ความสุ่ม” (creativity) ของคำตอบ               | - **ต่ำ (≈ 0-0.3)** → คำตอบคาดเดาได้ คล้ายเดิมทุกครั้ง  <br>- **กลาง (≈ 0.4-0.7)** → สมดุลระหว่างแม่นยำกับหลากหลาย  <br>- **สูง (≈ 0.8-2.0)** → คำตอบหลากหลาย/พลิกแพลงมากขึ้น | 1.0                                      |
| `max_tokens`  | กำหนด **จำนวนโทเคน** สูงสุดที่โมเดลสามารถสร้างในคำตอบ | - **น้อย** → คำตอบสั้น กระทัดรัด  <br>- **มาก** → คำตอบยาว รายละเอียดเยอะ (แต่ใช้เครดิตมากขึ้น)                                                                               | แตกต่างตามรุ่นโมเดล (เช่น 16k หรือ 128k) |

### สรุปสั้น ๆ

* `temperature` เหมือน “ปรับรสชาติ” — อุณหภูมิต่ำ = รสเรียบ, อุณหภูมิสูง = รสจัดหลากหลาย
* `max_tokens` คือ “ข้อจำกัดความยาว” ของคำตอบ ยิ่งมากยิ่งยาว (และใช้โควตา API มากขึ้น)

> **เคล็ดลับ**
>
> * ถ้าต้องการคำตอบทางเทคนิคหรือโค้ดที่แม่นยำ ใช้ `temperature` ต่ำ ๆ (0 – 0.3)
> * ถ้าอยากได้ไอเดียสร้างสรรค์ เช่น สตอรี่หรือสูตรอาหารใหม่ ๆ เพิ่ม `temperature` (≥ 0.8)
> * ปรับ `max_tokens` ให้พอดีกับความยาวที่ต้องการ เพื่อลดค่าใช้จ่ายและเวลารอผล


```python
user_prompt = "ขอวิธีทำข้าวผัดกระเพราไก่หน่อยครับ"

completion = openai_client.chat.completions.create(
    model=model_name ,
    messages=[
        {"role": "user", "content": user_prompt},
    ],
    
    temperature=0.8,
    max_tokens=400,
)

response = completion.choices[0].message.content
print(response)
```

```python

examples = [
    "บอกวิธีเลี้ยงแมว 7 ตัว",
    "เล่าประวัติของกรุงเทพมหานคร",
    "วิธีการทำอาหารไทยยอดนิยม",
    "แนะนำสถานที่ท่องเที่ยวในประเทศไทย",
    "เขียนบทความเกี่ยวกับการดูแลสุขภาพจิต"
]
examples
```

# test for loop

```python
for index, question in enumerate(examples):
    print(f"Question {index + 1}: {question}")
```

```python
for index, question in enumerate(examples):

    print(f"Question {index + 1}: {question}")

    
    completion = openai_client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "user", "content": question},
        ],
        temperature=0.8,
        max_tokens=400,
    )   
    response = completion.choices[0].message.content
    print(f"Answer {index + 1}: {response}\n")  
    print("-" * 50)  # Separator for clarity  
```
