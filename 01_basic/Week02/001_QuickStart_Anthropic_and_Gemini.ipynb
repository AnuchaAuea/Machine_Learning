# Define Message

```python
 # Define the messages for the conversation
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content":"ถ้ามนุษย์สามารถหยุดเวลาได้หนึ่งชั่วโมงต่อวัน คิดว่าจะเกิดผลกระทบต่อสังคมอย่างไรบ้างในระยะยาว?"}
]

messages 
```

# Open AI

```python
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)

answer = response.choices[0].message.content

display(Markdown(answer))
```

# Anthropic

```python
from anthropic import Anthropic
from IPython.display import display, Markdown

claude = Anthropic()

# Model name
model_name = "claude-sonnet-4-20250514"

# Create the API call with system parameter separate from messages
response = claude.messages.create(
    model=model_name,
    system="You are a helpful assistant.",  # System message goes here as a parameter
    messages=[
        {"role": "user", "content": "ถ้ามนุษย์สามารถหยุดเวลาได้หนึ่งชั่วโมงต่อวัน คิดว่าจะเกิดผลกระทบต่อสังคมอย่างไรบ้างในระยะยาว?"}
    ],
    max_tokens=1000
)

# Extract the text from the response
answer = response.content[0].text

# Display the answer
display(Markdown(answer))
```

# Gemini

```python
gemini = OpenAI(api_key=google_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
model_name = "gemini-2.0-flash"

response = gemini.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
```
